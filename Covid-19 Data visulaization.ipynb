```python
dataset.set_index('day') #x-axis
fig = plt.figure()
x1 = fig.add_subplot(111)
dataset['deaths'].plot(ax=x1 ,label='deaths')
plt.xlabel('day')
x1.set_ylabel('deaths')
plt.legend(loc=1)

x2= x1.twinx()
dataset['cases'].plot(ax=x2 , label='cases', color='green')
x2.set_ylabel('cases')
plt.legend(loc=2)
plt.show()
```
