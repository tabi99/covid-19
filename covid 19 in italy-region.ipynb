```python
#keeping only required columns
dataset = dataset[['Date','TotalHospitalizedPatients','Recovered','Deaths','TotalPositiveCases']]
```

```python
dataset
```

```python
X = dataset.drop('Deaths',axis=1)
y = dataset.Deaths
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=50)
```

```python
# For making Model 
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
acc =lr.score(X_test,y_test)
print('acc:',int(acc*100),'%')
```

```python
y_prd=lr.predict(X_test)
```

```python
df = pd.DataFrame({'Orginal':y_test, 'Predict':y_prd})
```
